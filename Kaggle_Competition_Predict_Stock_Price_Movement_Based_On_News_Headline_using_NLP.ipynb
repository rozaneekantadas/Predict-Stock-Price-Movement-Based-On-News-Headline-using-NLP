```python
import pandas as pd
import numpy as np
```

```python
file_path = '/content/Data.csv'
```

```python
df = pd.read_csv(file_path, encoding='ISO-8859-1')
```

```python
df.head()
```

```python
x = df.iloc[:,2:27]
y = df['Label']
```

```python
x.head(10)
```

```python
y.head(10)
```

```python
index_list = [i for i in range(25)]
str_index = [str(i) for i in index_list]
x.columns = str_index
x.head(10)
```

```python
x.replace("[^a-zA-Z]", " ", regex=True, inplace=True)
x.head()
```

```python
for i in str_index:
  x[i] = x[i].str.lower()
x.head()
```

```python
headlines = []
for row in range(0, len(x.index)):
  headlines.append(' '.join(str(lines) for lines in x.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
cv = CountVectorizer(ngram_range=(2,2))
x_data = cv.fit_transform(headlines)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y, test_size = 0.20, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
```

```python
classifier.fit(x_train, y_train)
```

```python
y_preds = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_pred=y_preds, y_true=y_test)
```

```python
cm
```

```python
classifier.score(x_test, y_test)
```

```python
print(classification_report(y_pred=y_preds, y_true=y_test))
```
